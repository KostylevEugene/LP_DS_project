```python
data['DATE'] = pd.to_datetime(data['DATE'])
```

```python
df1 = (data.set_index('DATE')
         .groupby('COMPANY')
         .apply(lambda x: x.asfreq('D', method='ffill'))
         .reset_index(level=0, drop=True)
         .reset_index()
         .reindex(data.columns, axis=1))
```
